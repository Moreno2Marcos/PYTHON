## **AMOSTRAGEM SISTEMÁTICA**

```
# Importa as dependencias
import numpy as np
import pandas as pd
```

```
# Monta o drive
from google.colab import drive
drive.mount('/content/drive')
```

```
# carrega os dados
enem_sp = pd.read_csv('/content/drive/MyDrive/UDEMY - Estatística para Análise de Dados/Moreno Marcos/microdados_enem_2019_sp_tratado.csv',
                      sep=',', encoding='iso-8859-1')
enem_sp.head()
```

```
# filtra os dados
enem_campinas = enem_sp.loc[enem_sp.NO_MUNICIPIO_RESIDENCIA == 'Campinas']
# reseta os índices para ordenação numérica
enem_campinas.reset_index(inplace=True)
enem_campinas
```

O cojunto de dados (população Campinas) são ordenamors numericamente pelos seus índices. A regra da sequência numérica seria: selecionar o 1ero elemento (aleatoriamente) do cojunto de dados com incrementos de 100 em 100.

```
# fixa a seleção aleatória
np.random.seed(2)
# escolhe aleatoriamente o primeiro elemento da amostra no intervalo 0 - 9 (pq?)
inicio = np.random.choice(10, 1)
inicio
```

```
# cria o conjunto de índices para seleção posterior 
indices1 = np.arange(inicio, len(enem_campinas), 100)
indices1
```

```
# visualiza tamanho do conj. de índices
len(indices1)
```

Utilizamos o cojunto de índices anterior criado para selecionar sistematicamente a amostra da população Campinas.

```
amos_sistematica = enem_campinas.iloc[indices1,:]
amos_sistematica
```
