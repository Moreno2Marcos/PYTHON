Iremos selecionar apenas a cidade de Sorocaba

```
enem_sorocaba = enem_sp.loc[enem_sp.NO_MUNICIPIO_RESIDENCIA == 'Sorocaba']
enem_sorocaba.head()
```

```
enem_sorocaba.shape
```

Agora selecionamos a coluna IDADE para análise

```
idades = enem_sorocaba['IDADE']
idades
```

Criamos uma variável de Frequência Absoluta associada às idades dos alunos particapantes do Enem que moram na cidade de Sorocaba. esta variável contém as contagem por idades dos participantes. 

Por exemplo. 8 alunos de 16 anos de idades participaram do ENEM na cidade de Sorocaba

```
freq_absoluta = Counter(idades)
freq_absoluta
```

Agora transformamos a variável freq_absoluta em um tipo de objeto DataFrame.

```
freq_absoluta = pd.DataFrame.from_dict(freq_absoluta, orient='index')
freq_absoluta
```

Vamos organizar os Indices do DataFrame Anterior. Assim, temos as idades dos alunos que participaram do Enem da cidade de Sorocaba.

```
freq_absoluta.sort_index(ascending=True, inplace=True)
freq_absoluta
```

Agora iremos criar a Frequência Relativa da IDADE dos participantes.

```
freq_relativa = freq_absoluta / freq_absoluta.sum()
freq_relativa
```

Agora calculamos a Frequec. Percentual Relativa e ordenamos de maior a menos porcentagem.

```
freq_relativa_perc = freq_relativa * 100
freq_relativa_perc.sort_values(by=0, ascending=False)
```

Agora faremos o mesmo para a cidade de Carapicuíba.

```
# filtra os dados, seleciona a coluna idade 
carapicuiba = enem_sp.query('NO_MUNICIPIO_RESIDENCIA == "Carapicuíba"')['IDADE']
# calcula a freq. absoluta
carapicuiba_freq_abs = carapicuiba.value_counts()
# calcula a freq. relativa
carapicuiba_freq_rel = carapicuiba_freq_abs / carapicuiba_freq_abs.sum()
# calcula a freq. precentual relativa
carapicuiba_freq_per_rel = carapicuiba_freq_rel * 100
carapicuiba_freq_per_rel.sort_values(ascending=False)
```
