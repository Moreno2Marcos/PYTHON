**Matplotlib**

```
import matplotlib.pyplot as plt
```

Selecionamos a variável IDADE dos particapentes do ENEM na cidade de Sorocaba. Analisaremos a sua distribuição de dados das IDADES. Repare o eixo Y que é a Frequencia Absoluta.

```
enem_sorocaba.IDADE.hist(bins=9)
plt.title('Distribuição de Idades - Enem SP 2019 - Sorocaba')
plt.show()
```

Observar no Histograma que idades menores a 20 anos possuem uma maior concetração. Ou seja, exsitiram mais participantes (por volta de 6000) na prova do ENEM do que as outras idades. Seguido isso, a faixa entre 20 e 30 anos.

Outra forma de plotar o Histrograma.

```
plt.hist(enem_sorocaba.IDADE, bins=8, rwidth=1)
plt.title('Distribuição de Idades - Enem SP 2019 - Sorocaba')
plt.show()
```

Especificando mais o Histograma. Repare no eixo Y que é a Densidade de Probabilidade? ou Freq. Relativa?

```
from matplotlib import colors
plt.hist(enem_sorocaba.IDADE, bins=8, rwidth=1, color='red', density=True)
plt.title('Distribuição de Idades - Enem SP 2019 - Sorocaba')
plt.xlabel('Idade')
plt.ylabel('Frequência')  # densidade de probabilidade?
plt.show()
```

Agora iremos criar o Histograma da nota da componente 1 da prova de redação dos estudantes de Sorocaba. Essa nota representa o domínio da modalidade escrita formal da língua portuguesa.

```
plt.hist(enem_sorocaba.COMP1, bins=10, rwidth=1, color='green', density=False)
plt.title('Histograma das Notas da Componente 1 da prova de Redação - Sorocaba')
plt.xlabel('Componente 1')
plt.ylabel('Frequência Absoluta')
plt.show()
```

Percebe-se no Histograma que aproximadamente na faixa de nota de 120 a 135 existiram uma maior quantidade de candidatos quando comparado a outras faixas de notas.

```
plt.hist(enem_sorocaba.COMP1, bins=10, rwidth=1, color='green', density=True)
plt.title('Histogram das Notas da Componente 1 da prova de Redação - Sorocaba')
plt.xlabel('Componente')
plt.ylabel('Frequência Relativa ou Densidade de Probabilidade?')
plt.show()
```

**Plotly**

Observe que no eixo Y é a Freq. Percentual Relativa

```
import plotly.express as px
grafico = px.histogram(enem_sorocaba, x="IDADE", nbins=10, histnorm='percent') 
grafico.update_layout(width=500, height=500, title_text='Distribuição de idades - Enem SP 2019 - Sorocaba') 
grafico.show()
```

Observe no Histograma anterior que entre 10 a 19 anos de idade concentra-se a maior quantidade de candidatos no ENEM 2019 aprox 67%. Diferentemente acontece na faixa de 70 -79 anos com 0.024%

Perceba no Eixo Y a contagem (Freq. Absoluta).

```
grafico = px.histogram(enem_sorocaba, x="IDADE", nbins=10, histnorm='density') 
grafico.update_layout(width=500, height=500, title_text='Distribuição de idades - Enem SP 2019 - Sorocaba') 
grafico.show()
```

No histograma anterior, aprox 543 alunos que aplicaram no ENEM encontram-se na faixa etária de 10 a 19 anos.

```
grafico = px.histogram(enem_sorocaba, x="IDADE", nbins=10, histnorm='probability') 
grafico.update_layout(width=500, height=500, title_text='Distribuição de idades - Enem SP 2019 - Sorocaba') 
grafico.show()
```

Perceba no gráfico anterior que o eixo Y representam a Freq. Relativa.

```
grafico =  px.histogram (enem_sorocaba,  x = "IDADE", nbins=10, histnorm='probability', color='SEXO') 
grafico.update_layout(width=600, height=500, title_text='Distribuição de idades por sexo - Enem SP 2019 - Sorocaba',
                      title_font_color = 'red') 
grafico.show()
```

No Histograma anterior, pode-se perceber que tanto a porcentagem de homens como de mulheres que aplicaram no ENEM 2019 são as mesmas independentemente da faixa etária.

Vamos agora a comparar a nota de matemática MT entre mulhere e homens.

```
grafico =  px.histogram(enem_sorocaba,  x = "NOTA_MT", nbins=20, color='SEXO') 
grafico.update_layout(width=500,height=500,title_text='Notas de Matemática por sexo - Enem SP 2019') 
grafico.show()
```

No Histograma de acima, temos:
- na Faixa de notas [25 - 74.9], houveram 437 mulheres e 298 homens que tiraram notas dentro dessa faixa;
- Na faixa de [425 - 474.9], teve-se a maior quantidade de notas dentre todas as faixas de notas. Onde a qtde de mulheres foi de 803 e de homens 403;
- No geral, têm uma maior quantidade de mulheres na faixa de [375 - 624] e na faixa de [624 - 1000] os homens são maioria.

```
grafico =  px.histogram (enem_sorocaba,  x = "NOTA_CN", nbins=20, color='SEXO', histnorm='percent') 
grafico.update_layout(width=500,height=500,title_text='Notas de Ciências Naturais por sexo - Enem SP 2019') 
grafico.show()
```

No histograma anterior, praticamente tanto as mulheres como os homens e número são equivalentes em todas as faixas de notas para Ciencias NAturais.

```
grafico =  px.histogram (enem_sorocaba,  x = "NOTA_LC", nbins=20, color='SEXO', histnorm='percent') 
grafico.update_layout(width=500,height=500,title_text='Notas de Linguagens e Códigos por sexo - Enem SP 2019') 
grafico.show()
```

No histograma anterior, praticamente (em termos de porcentagem) tanto as mulheres como os homens e número são equivalentes em todas as faixas de notas para Linguagem e Códigos.

```
grafico =  px.histogram (enem_sorocaba,  x = "NOTA_CH", nbins=20, color='SEXO') 
grafico.update_layout(width=500,height=500,title_text='Notas de Ciências Humanas por sexo - Enem SP 2019') 
grafico.show()
```

Na faixa de notas [375 - 574.9] existe uma maior quantidade das mulheres do que os homens na prova de Ciências Humanas. Perceba-se que SEXO é uma variável categórica, logo se faz uma análise entre uma variável numérica com 1 categórica num gráfico 2D.

```
grafico =  px.histogram (enem_sorocaba,  x = "NOTA_REDACAO", nbins=20, color='SEXO', histnorm='percent') 
grafico.update_layout(width=500,height=500,title_text='Notas de redação por sexo - Enem SP 2019') 
grafico.show()
```

Em linhas gerais, tanto os homens quanto as mulheres são equivalentes em porcentagens em todas as faixas de notas da Redação.

```
grafico =  px.histogram (enem_sorocaba,  x = "NOTA_REDACAO", nbins=20, color='RACA', histnorm='percent') 
grafico.update_layout(width=800,height=500,title_text='Notas de redação por raça - Enem SP 2019') 
grafico.show()
```

No geral, todos os tipos de raça não teria correção com a nota da Redação, já que, existe uma porcentagem equivalentes de cada tipo de raça em todas as faixas de notas. Porém, analisando mais de perto a faixa de maior nota [950 - 990] a parda amarela possui 0.2% do total.
