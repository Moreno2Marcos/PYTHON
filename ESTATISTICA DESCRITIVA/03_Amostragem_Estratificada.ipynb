```
# carrega os dados
enem_sp = pd.read_csv('/content/drive/MyDrive/UDEMY - Estatística para Análise de Dados/Moreno Marcos/microdados_enem_2019_sp_tratado.csv',
                      sep=',', encoding='iso-8859-1')
enem_sp.head()
```

Utilizaremos apenas os dados associados a CAMPINAS

```
# filtra os dados
enem_campinas = enem_sp.loc[enem_sp.NO_MUNICIPIO_RESIDENCIA == 'Campinas']
# reseta os índices para ordenação numérica
enem_campinas.reset_index(inplace=True, drop=True)
enem_campinas
```

Vamos considerar o atributo RACA para gerar os grupos que iremos separar.

```
enem_campinas['RACA'].value_counts()
```

Primeiro filtramos por um valor da raça (Ex. branca) e logo selecionamos aleatoriamente uma amostra (Ex. 10%)

**Para quantidade pequena de grupos**

```
# define tipos de raca
racas = enem_campinas['RACA'].value_counts().index
# define fração de seleção
frac = .10
# define lista de grupos
grupos = []
# itera cada tipo de raca para seleção
for raca in racas:
    grupos.append(enem_campinas.loc[enem_campinas.RACA == raca].sample(frac=frac))
# concatenamos cada grupo numa única tabela
amos_estra = pd.concat(grupos)
amos_estra.reset_index(drop=True, inplace=True)
amos_estra
```

Seleção estratificada final

```
amos_estra['RACA'].value_counts()
```

**Grandes quantidades de grupos**

```
enem_campinas['NO_MUNICIPIO_NASCIMENTO'].value_counts().index.size
```

```
# define nomes dos grupos
nomes = enem_campinas['NO_MUNICIPIO_NASCIMENTO'].value_counts().index
# define fração de seleção
frac = .10
# declara variável para armazenar seleção
grupos = []
# itera cada grupo para seleção
for nome in nomes:
    grupos.append(enem_campinas.loc[enem_campinas.NO_MUNICIPIO_NASCIMENTO == nome].sample(frac=frac))
# concatenamos a amaostra de cada grupo numa única tabela
amos_estra = pd.concat(grupos)
amos_estra.reset_index(drop=True, inplace=True)
amos_estra['NO_MUNICIPIO_NASCIMENTO'].value_counts()
```
