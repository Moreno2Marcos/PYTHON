Utilizaremos apenas os dados associados a CAMPINAS

```
# filtra os dados
enem_campinas = enem_sp.loc[enem_sp.NO_MUNICIPIO_RESIDENCIA == 'Campinas']
# reseta os índices para ordenação numérica
enem_campinas.reset_index(inplace=True, drop=True)
enem_campinas
```

Agora iremos a considerar apenas escolas públicas em Campinas.

```
escolas_publicas = enem_campinas.loc[enem_campinas.ESCOLA == 'pública']
escolas_publicas.reset_index(drop=True, inplace=True)
escolas_publicas.shape
```

```
escolas_publicas
```

Observe que a coluna CO_ESCOLA póssui valores faltantes.

```
escolas_publicas['CO_ESCOLA'].isnull().sum()
```

No entanto, isso não é problema porque o método value_counts() desconsidera valores faltantes na sua contagem.

```
amostras = escolas_publicas['CO_ESCOLA'].value_counts()
amostras.isnull().sum()
```

Criamos uma variável que vai armazenar os códigos das escolas públicas como sendo os grupos. Perceba a quantidade de alunos por grupo.

```
amostras = escolas_publicas['CO_ESCOLA'].value_counts()
amostras
```

Podemos acessar aos código das escolas da seguinte forma. Repare o número de escolas públicas.

```
amostras.index
```

Agora vamos a criar uma variável que armazene a seleção aleatória (amostragem). Repare que o tamanho deve ser igual ao tamanho da quantidade de escolas (como mostrada acima). Além disso, o critério de selção de grupos é 15% (do total de grupos) -  representado com o número "1".

```
np.random.seed(1)
amostragem = np.random.choice(a=[0,1], size=amostras.index.size, replace=True, p=[0.85, 0.15])
amostragem
```

Agora usando a amostragem anterior, iremos a selecionar os grupos (variável amostras) e armazenar numa variável para uso posterior.

```
amos_conglomerado = amostras.loc[amostragem == 1]
amos_conglomerado
```

```
amos_conglomerado.shape
```

```
amos_conglomerado.index
```

Finalmente iremos pegar os valores da tabela que considera escola públicas em Campinhas usando o indices anteriores (variável amos_conglomerado).

```
colegios = amos_conglomerado.index
amos_conglomerado_final = escolas_publicas.loc[escolas_publicas['CO_ESCOLA'].isin(colegios)]
amos_conglomerado_final.reset_index(drop=True, inplace=True)
amos_conglomerado_final
```
