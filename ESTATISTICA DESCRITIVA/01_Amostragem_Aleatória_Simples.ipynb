## **AMOSTRAGEM ALEATÓRIA SIMPLES**

Criamos uma semente aleatória para que toda vez que se execute a função SAMPLE retorne a mesma seleção de dados

```
np.random.seed(1)
```

Queremos selecionar aleatoriamente 20% do total de dados na cidade de Campinas

```
amos_simples = enem_campinas.sample(frac=0.20)  # , random_state=1
amos_simples.head()
```

A seleção aleatória também pode ser definada por uma quantidade fixa inteira

```
amos_simples = enem_campinas.sample(n=3000)  #, random_state=1
amos_simples.head()
```

Outra maneira de se fazer uma amostragem aleatória simples:

```
# fixa seleção aleatória
np.random.seed(2)
# cria os tamanhos de seleção de dados
amostra = np.random.choice(a=[0,1], size=len(enem_campinas), replace=True, p=[0.8, 0.2])
# visualiza o tamanho (80% do total de dados de Campinas)
print('80% :',len(amostra[amostra == 0]))
# visualiza o tamanho (20% do total de dados de Campinas)
print('20% :',len(amostra[amostra == 1]))
```

```
# seleciona aleatoriamente o 20% dos dados de Campinas
amos_simples = enem_campinas.loc[amostra == 1]
amos_simples
```
