## 1- Top 10 atores/atrizes considerando todos os dados

```python
# importamos a biblioteca pandas
import pandas as pd
```

```python
# carregamos dados do Netflix e observamos a coluna CAST
netflix = pd.read_csv('netflix_titles.csv')
netflix.head(2)
```

```python
# carregamos dados da Amazon e verificamos que possuem a mesma estrutura
amazon_prime = pd.read_csv('amazon_prime_titles.csv')
amazon_prime.head(2)
```

```python
# concatenamos os dados da Amazon e Netflix
all_data = pd.concat([netflix, amazon_prime], ignore_index=True)
# acessamos à coluna CAST
cast = all_data['cast']
# filtramos os valores vazios(NaN) da coluna CAST
cond = cast.notnull()
cast = cast[cond]
# resetamos os indices
cast.reset_index(inplace=True, drop=True)
# convertemos os seus valores a minusculo
cast = cast.str.lower()
cast.head(3)
```

```python
# transformamos a coluna CAST numa matriz (DataFrame) cujos valores são os nomes dos atores/atrizes
cast_df = cast.str.split(',', expand=True)
cast_df.head(2)
```

```python
# OBJETIVO: criar uma única coluna contendo os nomes dos atores/atrizes para fazer a contagem

# capturamos a primiera coluna
col = cast_df[0]

# percorremos cada coluna da matriz
for i in range(1, 76):
    # concatenamos una vez só a primeira coluna com a segunda
    if i == 1:
        result = pd.concat([col, cast_df[i]], ignore_index=True)
    # concatenamos a coluna resultante com as restantes e sobre-escrevemos nela mesma  
    else:
        result = pd.concat([result, cast_df[i]], ignore_index=True)

# filtramos valores vazios (None)
cond = result.notnull()
result = result[cond]
# removemos espaços em branco
result = result.str.strip()
# listamos os 10 nomes de atores/atrizes que mais se repetem
result.value_counts().head(10)
```
