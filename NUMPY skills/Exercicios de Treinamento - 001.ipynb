## **P3 - Quais são os nomes dos 3 melhores alunos no "assignment1"**?

```python
# mostra os dados
notasAlunos
```

```python
# seleciona os históricos dos alunos com apenas as colunas students e assigment1
melAlus = []
for aluno in notasAlunos:
    melAlus.append(list(aluno[1:3]))
melAlus
```

```python
# ordena os históricos segundo a nota
melAlus.sort(key=lambda x: x[1],  reverse=True)
melAlus
```

```python
# captura o histórico doss 3 melhores alunos
melAlus[:3]
```

```python
# seleciona o nome dos 3 melhores alunos
for item in melAlus[:3]:
    print(item[0])
```

## **P3 - Quais são os nomes dos 3 piores alunos no "assignment3"**?

```python
# mostra os dados
notasAlunos
```

```python
# captura o histórico de todos os alunos considerando apenas a coluna "student" e "assignment3"
pioAluns = []
temp = []
for aluno in notasAlunos:
    temp.append(aluno[1])
    temp.append(aluno[4])
    pioAluns.append(temp)
    temp = []
pioAluns
```

```python
# ordena os históricos segundo a nota
pioAluns.sort(key=lambda x: x[1],  reverse=False)
pioAluns
```

```python
# seleciona o nome dos 3 piores alunos
for item in pioAluns[:3]:
    print(item[0])
```
