# **EXERCÍCIO - 01**

Criar duas listas, uma que represente as quantidades de vendas de vários produtos e outra que represente seus preços de custos unitários respectivos. Calcule o custo total de cada produto.

### **RESOLUÇÃO**

```python
from numpy import array
from random import randint

# cria uma lista de 5 quantidades aleatórios entre 1 e 10
qtds = [randint(1, 10) for i in range(5)]
# converte para array as quantidades
qtdsArr = array(qtds)
qtdsArr
```

```python
# cria uma lista de 5 preços aleatórios entre 50 e 100
preUnits = [randint(50, 100) for i in range(5)]
# converte para array os peços
preUnitsArr = array(preUnits)
preUnitsArr
```

```python
# calcula o custo total
cusTotArr = qtdsArr * preUnitsArr
cusTotArr
```

Se agora temos o preço de venda de cada um desses produtos na lista:

```python
preVends = array([randint(60, 110) for i in range(5)])
preVends
```

Calcula o lucro total de cada produto:

```python
lucroTots = (qtdsArr * preVends) - cusTotArr
lucroTots
```

# **EXERCÍCIO - 02**

Criar uma sequência de números inteiros no intervalo [1 - 20] com incrementos de 2 em 2.

### **RESOLUÇÃO**

```python
from numpy import arange

arange(start=1, stop=20, step=2, dtype=int)
```

# **EXERCÍCIO - 03**

Criar uma sequência de 7 números inteiros no intervalo [1 - 20] igualmente espaçado.

### **RESOLUÇÃO**

```python
from numpy import linspace

linspace(start=1, stop=20, num=7, dtype=int)
```
